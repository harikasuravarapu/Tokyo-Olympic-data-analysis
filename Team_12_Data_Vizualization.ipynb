```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming it's stored in a CSV file)

# Specify the file path
file_path = r"C:\Users\harik\Downloads\data\Athletes.csv"

# Load the dataset using a different encoding
olympic_data = pd.read_csv(file_path, encoding='latin1')

# Visualizing Number of athletes per discipline
plt.figure(figsize=(10, 10))
sns.countplot(x='Discipline', data=olympic_data)
plt.xticks(rotation=45)
plt.xlabel('Discipline')
plt.ylabel('Number of Athletes')
plt.title('Number of Athletes per Discipline')
plt.tight_layout()
plt.show()

# visualizing Number of athletes per team
plt.figure(figsize=(10, 10))
sns.countplot(x='Country', data=olympic_data)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Athletes')
plt.title('Number of Athletes per Team')
plt.tight_layout()
plt.show()

# visualization: Distribution of medals by country
plt.figure(figsize=(12, 8))
sns.countplot(x='Discipline', hue='Country', data=olympic_data)
plt.xlabel('Discipline')
plt.ylabel('Number of Medals')
plt.title('Distribution of Medals by Country')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
file_path = r"C:\Users\harik\Downloads\data\Coaches.csv"
coaches_data = pd.read_csv(file_path, encoding='latin1')

# Assuming coaches_data is your DataFrame containing the coaches dataset
country_counts = coaches_data['Country'].value_counts()
country_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Number of Coaches')
plt.title('Number of Coaches by Country')
plt.show()
```

```python
event_counts = coaches_data['Event'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(event_counts, labels=event_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Event Distribution for Coaches')
plt.axis('equal')
plt.show()
```

```python
file_path = r"C:\Users\harik\Downloads\data\EntriesGender.csv"
entriesgender_data = pd.read_csv(file_path, encoding='latin1')

entriesgender_data.set_index('Discipline')[['Female', 'Male']].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Discipline')
plt.ylabel('Number of Entries')
plt.title('Female and Male Entries by Discipline')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

entriesgender_data.set_index('Discipline')[['Female', 'Male']].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Discipline')
plt.ylabel('Number of Entries')
plt.title('Female and Male Entries by Discipline')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
plt.bar(entriesgender_data['Discipline'], entriesgender_data['Total'], color='skyblue')
plt.xlabel('Discipline')
plt.ylabel('Total Entries')
plt.title('Total Entries by Discipline')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
